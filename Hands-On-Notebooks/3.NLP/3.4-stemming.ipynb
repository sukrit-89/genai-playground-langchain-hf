Stemming 

Stemming is the process of reducinga word stem that fixes that affixes and suffixes and prefixesor to the roots of the words known as lemma. Stemming is important in natural language understanding (NLU) and natural language processing(NLP)

```python
##classification problem

#comment of the product is a positive review or neagative review
#Reviews-->eating,eat,eaten {going,gone,goes}--> here eat and go are the stem of them

#so finding this word stem is done using stemming

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]

#porter stemmer-->Technique of stemming

from nltk.stem import PorterStemmer
stemming=PorterStemmer()

for word in words :
    print(word+"---->"+stemming.stem(word))
 
```

Major disadvantages of stemming:
For many words it cannot give a proper meaningful word

for example if we see history-->histori (meaningless)

```python
#examples:

stemming.stem('congratulations')
##here we see agin it give a meaningless word 'congratul' as a word stem of congratulations
```

RegexpStemmer class

NlTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression

```python
from nltk import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
reg_stemmer.stem('eating')
#here the $ specifies that the expressions are only removed from the end of the word
```

```python
reg_stemmer_n=RegexpStemmer('ing|s$|e$|able$',min=4)
reg_stemmer_n.stem('ingeating')
```

Snowball Stemmer --> stemming techiniques gives a better accuracy than porter stemmer(accuaracy in the sense better words)

snowball stemmer supports Arabic,Danish,Dutch,English,French,German and many more...

```python
from nltk.stem import SnowballStemmer
snb_stemmer=SnowballStemmer('english')
for word in words:
    print(word+"---->"+snb_stemmer.stem(word))

```

Differce bw porter and snowball lets check
