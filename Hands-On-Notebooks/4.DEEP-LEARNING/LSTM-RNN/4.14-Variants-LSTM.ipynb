# Variants of LSTM RNN

The base architecture is similar to the LSTM but the difference is the connections ->
- From memory cell to forget gate, input gate, output gate 

- This connections are known as peephole connections

# Peephole connections

we let the gate layers look at the cell state and thus influence of cell state works in taking the desicion of keeping contexts.

Thus the accuray of forgeting, inputing and outputing operations increase

# Another variation

 Base architecture is similar to RNN but here we are coupling forget and input gate 

 Insted of sperately deciding what to forget and what should be added as new info , we make this decision together

 Goal: When we only forget when we are going to input something in its place 

## Conclusion: We only input new values to the state when we forget something older
