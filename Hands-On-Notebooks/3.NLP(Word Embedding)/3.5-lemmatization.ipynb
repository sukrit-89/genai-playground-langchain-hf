Lemmatization is like stemming. The output we will get after lemmatization is called 'lemma', which is root ward rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing

NLTK provides WordNetLemmatizer class which is a thin wrapper around wordnet corpus. This class uses morphy() function to thhe WordNet CorpusReader class to find lemma. Let us understand it with an example-

```python
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
lem.lemmatize('going',pos='n')
```

```python
'''
POS-
Noun -n
verb -v
adjective -a
adverb -r
'''
lem.lemmatize('going',pos='v')
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
for word in words :
    print(word+"---->"+lem.lemmatize(word,pos='v'))
    #by default-> noun -n
```

```python
print(lem.lemmatize('goes',pos='n'))
lem.lemmatize('fairly',pos='n'),lem.lemmatize('sportingly',pos='n')
```

Lemmatization is beter but significantly slower than stemming because it requires:

    - Dictionary lookup: It needs to access a comprehensive dictionary/corpus (like WordNet) to find the correct lemma
    - Part-of-speech analysis: For accurate results, it often requires determining the part of speech of each word first
    - Morphological analysis: It performs more complex linguistic analysis to understand the word's structure

This makes lemmatization much more resource-intensive and time-consuming compared to stemming, which uses simple rule-based suffix removal. For large-scale text processing applications or real-time systems, this performance overhead can be a significant limitation.

As you can see in your notebook examples, lemmatization requires specifying the POS tag (pos='v', pos='n') for optimal results, adding another layer of complexity that stemming doesn't require.
