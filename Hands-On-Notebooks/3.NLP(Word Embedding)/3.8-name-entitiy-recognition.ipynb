NAME ENTITY RECOGNITION

```python
sentence="The Eiffel Tower was built from 1887 to 1889 by French Engineer Gustav Eiffel, whose company specialized in building metal frameworks and structures."
'''
Name entity tags:

person Eg:Sukrit Goswami
Place or Location: India
Date Eg:31-01-2025
time Eg:9:30 pm
Money Eg: 1 Million Dollar
organization Eg: XYZ pvt. ltd.
Percent Eg: 20%, Twenty percent
'''
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
tag_elements=nltk.pos_tag(words)
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
nltk.ne_chunk(tag_elements).draw()
```
