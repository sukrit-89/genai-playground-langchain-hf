```python
corpus="Hello, I am Sukrit and i have interest in Ai-ML.I am BTech undergrad student"
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize
documents=sent_tokenize(corpus)
```

```python
for sentence in documents:
    print(sentence)
```

```python
#paragraphs-->words
#senctence-->words
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

For word punction tokenization

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

In TreeBankWordTokenizer full stop is not treated as a seperate word except the last full stop

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
