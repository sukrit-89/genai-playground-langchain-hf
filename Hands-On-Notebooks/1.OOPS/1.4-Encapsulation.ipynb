Encapsulation and Abstarction 

Encapsulation and Abstraction are two fundamental principles of OOP that helps in designing about robust,maintainable and reusable code.
Encapsulation involves bundling data and methods that operate on the data within a single unit, while abstraction involves hiding complex implementation details and exposing only the necessary features  

Encapsulation is the concept of wrapping (variables) and methods (functions) together as a single unit.It restricts direct access to some of the object's components,which is means of preventing accidental interference ans misuse of the data

```python
#Encapsulation with Getter and Setter methods
#public,protected,private varriables

class Person:
    def __init__(self,name,age):
        self.name=name #public variable
        self.age=age   #public variable
person=Person("Sukrit",19)
'''so now a outside function is getting the access'''

def get_name(person):
    return person.name      

get_name(person)
```

```python
#inbuilt methods or variables that are present
dir(person)
```

```python
#to prevent the acces of the varibale we need make them private
class Person:
    def __init__(self,name,age,height):
        self.__name=name #private variable
        self.__age=age   #private variable
        self.height=height
person=Person("Sukrit",19,5)
dir(person)
```

```python
#protected variable is similar to priavte variable but it can be used by a derived class
class Person:
    def __init__(self,name,age,height):
        self._name=name #protected variable
        self._age=age   #protected variable
        self.height=height

class Employee(Person):
    def __init__(self,name,age,height):
        super().__init__(name,age,height)
'''only accessible to derived class'''
emp=Employee("Sukrit",19,6)
print(emp._name)
```

```python
#Encapsulation using Getter and Setter Method

class Person:
    def __init__(self,name,age,height):
        self.__name=name #private access modifier or variable
        self.__age=age#private variable
    #Getter method for name
    def get_name(self):
        return self.__name
    #Setter method for name
    def set_name(self,name):
        self.__name=name
    #Getter method for age
    def get_age(self):
        return self.__age
    #Setter method for age
    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("Age cannot be negative")
#Access and modify the private variables with Getter and Setter method
person=Person("Sukrit",19,6)
print(person.get_name())
print(person.set_age(2))
print(person.get_age())
```
