Magic Methods

Magic methods in python also known as dunder menthod(double underscore methods),are special methods that start and end with double underscores.These methods enable you to define the behaviour of objects for built in operations, such as arithmetic operations,comparisons and more.

Magic Methods are predefined methods in python that you can override to change the behavior of your objects.

```python
'''Some common magic methods include:
- __init__: Initalizes a new instance of class
- __str__: Returns a string representaion of a object
- __repr__: Returns an official string representation of an object
- __getitem__: Gets an item from a container
-__setitem__: Sets an item in a container'''
```

The message is returned by the method __str__

```python
#Basics Magic Methods
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
person=Person("Sukrit",19)
print(person)
```

```python
'''Overriding the __str__ method'''
#Basics Magic Methods
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        '''Here the functionality of the __str__ method is chnaged'''
    def __str__(self):
        return f"{self.name},{self.age} years old"
    def __repr__(self):
        return f"(name={self.name},age={self.age})"
person=Person("Sukrit",19)
print(person)
print(repr(person))
```
