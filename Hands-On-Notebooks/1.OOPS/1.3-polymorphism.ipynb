Polymorphism

It is core concept in OOP that allows objects of differnt classes to be treated as objects of a common superclass.it provide a way to perform a single action in different forms.

Polymorphism is typically achieved through method overridding and interfaces

```python
#method overridding
class Animal:
    def speak(self):
        return "sound of animal"

#derived class
class Dog(Animal):
    def speak(self):
        return "woof"

class Cat(Animal):
    def speak(self):
        return "meow"
```

```python
dog=Dog()
print(dog.speak())
cat=Cat()
print(cat.speak())
```

```python
#function that demonstartes polymorphisim

def ani_speak(animal):
    print(animal.speak())
ani_speak(dog)
```

```python
#polymorphism with functions and methods
class Shape:
    def area(self):
        return "The area of the shape"
#derived class1
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height
    def area(self):
        return self.width*self.height
#derived class 2
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14*self.radius*self.radius

#function which demonstrate polymorphism
def print_area(shape):
    print(f"the area is {shape.area()}")
rectangle=Rectangle(4,5)
circle=Circle(5)
print_area(rectangle)
print_area(circle)
    
```

Polymorphism with Abstract Base Class
Abstarct Base Class (ABCs) are used to define common methods for a gorup of related objects.
they can enmforce taht derived classes implement particular methods,promoting consistency across differdnt implementation

```python
from abc import ABC,abstractmethod

class Vehicle(ABC):
    @abstractmethod 
    def start_engine(self):
        pass
    
#derived class1
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"
#derived class2
class Motorcycle(Vehicle):
    def start_engine(self):
        return "MotorCycle engine started"
def start_vehical(vehical):
    print(vehical.start_engine())

#create objects
car=Car()
motorcycle=Motorcycle()
start_vehical(car)
start_vehical(motorcycle)
```
