Excepting handling in python allows you to handle errors gtracefully and take corrective actions without stopping the execution of the program

Exceptions-> They are the events that disrupts the normal flow of the program.
They occur when an error is encountered during the program execution.

Common exceptions include:
1.ZeroDivisionError: Diving by zero
2.FileNotFoundError:File not found.
3.ValueError:Invalid value

```python
#example-1
a=b
```

```python
#fixing the above error
try:
    a=b
    '''here still the error is present by handled with try and except block '''
except NameError as ex:
    print(ex)
```
