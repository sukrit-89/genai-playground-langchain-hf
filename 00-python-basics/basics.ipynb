Example coding of a function

convert celcius to faranhite 

```python
def convert(temp,unit):
    if unit == 'c'or unit=='C':
        print("The temperature is converted into Fahrenheit")
        return temp*9/5+32
    elif unit == 'f' or unit=='F':
        print("The temperature is converted into celsius")
        return (temp-32)*5/9
    else:
        return None
    
```

```python
print(convert(25,'c'))
print(convert(77,'f'))
```

Lambda Function

Lamda functions are small anonyms functions defined using the lambda keyword.They can have any number of args but only one expression.
They are commonly used for short operations or as arguments to higher order functions

```python
#Example of lambda Function
addition=lambda a,b:a+b
print(type(addition))
addition(5,2)
```

Map Funuction -> applies to a function to all items in the list

```python
#Example using lambda

nums=[1,2,5,8,9,6]
test_1=list(map(lambda x:x**2,nums))
print(test_1)


def cube(x):
    return x**3
test_2=list(map(cube,nums))
print(test_2)
```

Filter function-->
constructs an iterator from elements of an iterable for which a function returns true.
It is used to filter out elements from a list based on condition

```python
#Example
def even(x):
    if x%2==0:
        return True
```

```python
lst=[2,54,8,5,6,32,4,24]

out=list(filter(even,lst))
print(out)
```
