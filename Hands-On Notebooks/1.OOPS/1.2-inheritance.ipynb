Inheritance in Python

Inheritance is a fundamental concept in Object-Oriented Programming(OOP) that allows a class to inherit attributes and methods from another class.

```python
#inheritance (Single)
#parent class
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
    def drive(self):
        print(f"The person will drive the {self.enginetype} car")
```

```python
car1=Car(4,5,"petrol")
print(car1.drive())
```

```python
#child class
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        '''super is method to inherit the parent class '''
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving=is_selfdriving
    def selfdriving(self):
        print(f"Tesla supports self driving:{self.is_selfdriving}")
```

```python
Tesla1=Tesla(4,5,'ev',False)
print(Tesla1.selfdriving())
print(Tesla1.drive())
```

```python
#Multiple Inheritance 
#when a class inherits more than one base class
class Anime:#base class1
    def __init__(self,name):
        self.name=name
    def watch(self):
        print(f"Bro is watching {self.name}")
class Anime_Writer:
    def __init__(self,writer):
        self.writer=writer


#derived class
class Overview(Anime,Anime_Writer):
    def __init__(self,name,writer):
        Anime.__init__(self,name)
        Anime_Writer.__init__(self,writer)
    def info(self):
        return f"{self.name} written by {self.writer}"
           
#object
OnePeice=Overview("AOT","Isyama")
print(OnePeice.info())
```
