Operator Overloading

Operator Overloading allows you to define the behavior of the operators (+,-,*,etc.)for custom objects. You acheive by overriding specific magic methods in your class

```python
### Common Operators Overloading Magic Methods
'''
__add__(slef,other): Adds two objects using the + operator 
__sub__(slef,other): Subtracts two objects using the - operator
__mul__(slef,other): Multiplies two objects using the * operator
__truediv__(slef,other): Divides two objects using the / operator
__eq__(self,other): checks if two objects are equal using the == operator
__lt__(self,other):checks if one object is less than another using the < operator
'''
```

Mathematical operations for vectors by operator overloading

```python
class Vector:
    def __init__(self,a,b):
       self.a=a
       self.b=b
    def __add__(self,other):
        return Vector(self.a+other.a,self.b+other.b)
    def __sub__(self,other):
            return Vector(self.a-other.a,self.b-other.b)
    def __mul__(self,other):
        return Vector(self.a*other.a,self.b*other.b)
    def __eq__(self,other):
        return Vector(self.a==other.a,self.b==other.b)
    
    def __repr__(self):
        return f"Vector({self.a},{self.b})"
    
#create objects
v1=Vector(2,3)
v2=Vector(4,5)
print(v1+v2)
print(v1*v2)
```

Conclusion-->
--Operator overloading allows custom classes to behave like built-in types by implementing magic methods.
--The Vector example shows how __add__, __sub__, __mul__, and __eq__ enable intuitive elementwise operations, and __repr__ makes results readable.
--Use these methods to make your objects clearer, more Pythonic, and easier to work with.
