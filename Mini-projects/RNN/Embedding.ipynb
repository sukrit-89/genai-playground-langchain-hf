Practical Implementation of Word Embedding 

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#Sentences

sent=['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'yot videos are good']
```

```python
sent
```

```python
#Define the vacab size
voc_size=10000
```

```python
#Onehot Representation

one_hot_rep=[one_hot(words,voc_size)for words in sent]
one_hot_rep #giving that at what index we will be gtting '1'
```

```python
## Word embedding Representation

from tensorflow.keras.layers import Embedding,Flatten
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential 
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length) # making the sentences of equal size 

print(embedded_docs)
```

```python
# feature rep

dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python

model.build(input_shape=(None, 8))

model.summary()
```

```python
model.predict(embedded_docs) # Every Word is represented by 10 dims
```
